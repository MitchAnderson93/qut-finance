### Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Define Data

```python
# Data for the assets
expected_return_A = 0.15  # 15%
std_dev_A = 0.20         # 20%
expected_return_B = 0.10  # 10%
std_dev_B = 0.15         # 15%
correlation = 0.3

# Covariance between Asset A and B
covariance = correlation * std_dev_A * std_dev_B
```

### Portfolio Calculations

```python
# Portfolio weights
weights = np.linspace(0, 1, 100)

# Portfolio returns
portfolio_returns = weights * expected_return_A + (1 - weights) * expected_return_B

# Portfolio variance and standard deviation
portfolio_variances = (
    weights**2 * std_dev_A**2 +
    (1 - weights)**2 * std_dev_B**2 +
    2 * weights * (1 - weights) * covariance
)
portfolio_std_devs = np.sqrt(portfolio_variances)
```

### Plot the Efficient Frontier

```python
# Plot
plt.figure(figsize=(10, 6))
plt.plot(portfolio_std_devs, portfolio_returns, label='Efficient Frontier', color='blue')
plt.scatter(std_dev_A, expected_return_A, color='red', label='Asset A', s=100)
plt.scatter(std_dev_B, expected_return_B, color='green', label='Asset B', s=100)

# Labels and legend
plt.title("Risk-Return Profile (Efficient Frontier)")
plt.xlabel("Risk (Standard Deviation)")
plt.ylabel("Expected Return")
plt.legend()
plt.grid(True)
plt.show()
```
